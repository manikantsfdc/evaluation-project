```python
#In every plot we can see that people with ae 55+ have more risk or more targeted.
a = sns.FacetGrid(heartdata, col = "target")
a.map(sns.distplot, "age", bins = 20)
plt.show()
```

```python
#0 for female and 1 is for male. We can see that female are more prone to target just between 0-2 whereas in case of men, less chnaces but can be seen in all target 0-4.
b= sns.FacetGrid(heartdata, col = "sex")
b.map(sns.distplot, "target", bins = 20)
plt.show()
```

```python
#According to dataset provided we can see more male members have probability of heart disease
b = sns.factorplot(x = "sex", y = "target", data = heartdata, kind = "bar",size=5)
b.set_ylabels("Probability of Heart Disease")
b.set_xlabels("Sex")
plt.show()
```
