```python
# we will convert categorical numerically encoded columns into their categories

# binning worklifebalance
def bin_work_life(n):
    if n ==1:
        return 'Bad'
    elif n ==2:
        return 'Good'
    elif n ==3:
        return 'Better'
    elif n ==4:
        return 'Best'

data['WorkLifeBalance'] = data['WorkLifeBalance'].apply(lambda x: bin_work_life(x))
```

```python
# we will convert categorical numerically encoded columns into their categories

# binning worklifebalance
def bin_work_life(n):
    if n ==1:
        return 'Bad'
    elif n ==2:
        return 'Good'
    elif n ==3:
        return 'Better'
    elif n ==4:
        return 'Best'

data['WorkLifeBalance'] = data['WorkLifeBalance'].apply(lambda x: bin_work_life(x))

# binningPerformanceRating
def bin_PerformanceRating(n):
    if n ==1:
        return 'Low'
    elif n ==2:
        return 'Good'
    elif n ==3:
        return 'Excellent'
    elif n ==4:
        return 'Outstanding'

data['PerformanceRating'] = data['PerformanceRating'].apply(lambda x: bin_PerformanceRating(x))

# binning JobSatisfaction
def bin_JobSatisfaction(n):
    if n ==1:
        return 'Low'
    elif n ==2:
        return 'Medium'
    elif n ==3:
        return 'High'
    elif n ==4:
        return 'Very High'

data['JobSatisfaction'] = data['JobSatisfaction'].apply(lambda x: bin_JobSatisfaction(x))

#education degree
def bin_Education(n):
    if n ==1:
        return 'Below College'
    elif n ==2:
        return 'College'
    elif n ==3:
        return 'Bachelor'
    elif n ==4:
        return 'Master'
    elif n==5:
        return 'Doctor'

data['Education'] = data['Education'].apply(lambda x: bin_Education(x))
```

```python
# encoding attrition to binary value 0(no attrition) and 1(attrition)
data['Attrition'] = data['Attrition'].apply(lambda x: 0 if x=='No' else 1)

# converting Attrition datatype to integer type
data['Attrition'] = data['Attrition'].apply(lambda x: pd.to_numeric(x))

# segregating dataframes based on class label for plotting in future

attr_1=data[data['Attrition']==1]
attr_0=data[data['Attrition']==0]
```

```python
# defining a function which takes a categorical variable and plots the Attrition rate
# segmented by Gender 
def plot_segmented(cat_var):
    plt.figure(figsize=(10, 6))
    sns.barplot(x=cat_var, y="Attrition", hue="Gender", data=data)
    plt.show()
```

```python
# Education by Gender Plotting
plot_segmented('Education')
#Observation : 1.Females are more as Doctor , college and below college education , whereas males are more in master and bachelors
```

```python
# Education Fields by Gender Plotting
plot_segmented('EducationField')
#Most Hr and life science students are Females
#More males in Technical degree
```
